```python
import numpy as np
from math import sin, cos, exp, pi
from matplotlib import pyplot as plt
```

```python
pts = 50
domain = np.linspace(0, 1, pts)
dx=1/pts
l = 20*dx
k=2*pi/l
```

```python
f =np.sin(k*domain)
plt.plot(domain, f, label='sin(kx)')
plt.show()
```

```python
nder =np.zeros(pts)
ader = k*np.cos(k*domain)
ader[0]= 0
ader[pts-1]= 0
for i in range(1,pts-1):
    nder[i] = (f[i+1]-f[i-1])/(2*dx)
error = np.sqrt(np.mean((nder-ader)**2))
```

```python
plt.plot(domain, nder, label='Numerical Derivative')
plt.plot(domain, ader, label='Analytical Derivative')   
plt.plot(domain,nder-ader, label='Error')
plt.legend()
plt.grid()
plt.show()
print(error)
```
